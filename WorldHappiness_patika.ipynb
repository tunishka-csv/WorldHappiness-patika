## Data Visualization

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="crest", fmt=".2f")
plt.title("Happiness Correlation Matrix")
plt.show()
```

Happiness seems to have strong correlation with all parameters.

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="GDP per capita", y="Score", data=df)
plt.title("GDP per Capita vs. Happiness Score")
plt.xlabel("GDP per Capita")
plt.ylabel("Happiness Score")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="Social support", y="Score", data=df)
plt.title("Social Support vs. Happiness")
plt.xlabel("Social Support")
plt.ylabel("Happiness Score")
plt.grid(True)
plt.show()
```

```python
top_10 = df.sort_values("Score", ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x="Score", y="Country or region", data=top_10, palette="viridis")
plt.title("Top 10 Happiest Countries")
plt.xlabel("Happiness Scores")
plt.ylabel("Country")
plt.show()
```

```python
bottom_10 = df.sort_values("Score", ascending=True).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x="Score", y="Country or region", data=bottom_10, palette="Reds")
plt.title("Top 10 Least Happy Countries")
plt.xlabel("Happiness Score")
plt.ylabel("Country")
plt.show()
```
